```python
# drop company
df.drop(['company', 'country', 'market_segment', 'distribution_channel'], axis=1, inplace=True)
# fill missing values with 0 in agent
df['agent'].fillna(0, inplace=True)
df['children'].fillna(0, inplace=True)
df['agent']
df.dtypes
# filling country
```

```python
#Prepare data ground truth and normalize the data
adr_val = df['adr'].to_numpy()
adr_mean, adr_std = adr_val.mean(), adr_val.std()
adr_normalized = (adr_val - adr_mean) / (adr_std)
#print(adr_mean, adr_std)
#print(adr_normalized.mean(), adr_normalized.std())
#print(adr_val.shape)
#prepare feature data
is_canceled = df['is_canceled'].to_numpy()
```

```python
# arrival date month
from sklearn import preprocessing
#le_month = preprocessing.LabelEncoder()
#le_month.fit(df['arrival_date_month'].to_numpy())
#le_month.classes_
#df['arrival_date_month'] = le_month.transform(df['arrival_date_month'].to_numpy())
#df['arrival_date_month'].value_counts()
#arrival_date_month
month_dict = {'January':1, 'February':2, 'March':3, 'April':4, 'May':5, 'July':7, 'June':6, 'August':8, 'September':9, 'October':10, 'November':11, 'December':12}
df.replace({'arrival_date_month':month_dict}, inplace=True)
#df['arrival_date_month'].value_counts()
```

```python
# hotel
le_hotel = preprocessing.LabelEncoder()
le_hotel.fit(df['hotel'].to_numpy())
le_hotel.classes_
df['hotel'] = le_hotel.transform(df['hotel'].to_numpy())
#df['hotel'].value_counts()
```

```python
# meal
le_meal = preprocessing.LabelEncoder()
le_meal.fit(df['meal'].to_numpy())
le_meal.classes_
df['meal'] = le_meal.transform(df['meal'].to_numpy())
#df['meal'].value_counts()
```

```python
#reserved_room_type, assigned_room_type
# 10, 12 ????
# use le_assigned to transform both reserved_room_type and assigned_room_type
le_assigned = preprocessing.LabelEncoder()
le_assigned.fit(df['assigned_room_type'].to_numpy())
#print(le_assigned.classes_)
#le_reserved = preprocessing.LabelEncoder()
#le_reserved.fit(df['reserved_room_type'].to_numpy())
#le_reserved.classes_
#reserved_room_type
df['reserved_room_type'] = le_assigned.transform(df['reserved_room_type'].to_numpy())
#print(df['reserved_room_type'].value_counts())
df['assigned_room_type'] = le_assigned.transform(df['assigned_room_type'].to_numpy())
#print(df['assigned_room_type'].value_counts())
```

```python
#deposit_type 
le_de = preprocessing.LabelEncoder()
le_de.fit(df['deposit_type'].to_numpy())
le_de.classes_
df['deposit_type'] = le_de.transform(df['deposit_type'].to_numpy())
#df['deposit_type'].value_counts()
```

```python
#customer_type
le_cus=preprocessing.LabelEncoder()
le_cus.fit(df['customer_type'].to_numpy())
le_cus.classes_
df['customer_type'] = le_cus.transform(df['customer_type'].to_numpy())
#df['customer_type'].value_counts()
```

```python
#reservation_status
le_re=preprocessing.LabelEncoder()
le_re.fit(df['reservation_status'].to_numpy())
le_re.classes_
df['reservation_status'] = le_re.transform(df['reservation_status'].to_numpy())
#df['reservation_status'].value_counts()
```
